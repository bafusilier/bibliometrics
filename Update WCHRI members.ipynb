Update `chw_ids.xlsx` with the new WCHRI auids.

```python
df = pd.read_excel('../../Datasets/wchri_with_ids_updated.xlsx')
df_chw = pd.read_excel('../../Datasets/chw_ids.xlsx')

df.info()
df_chw.info()
```

```python
to_add = df[['Name', 'Auid']]
to_add
```

```python
to_add[to_add['Auid'].isna()]
```

```python
# found one more missing researcher
to_add.loc[128, 'Auid'] = 57016314300
```

```python
to_add[to_add['Auid'].isna()]
```

```python
to_add = to_add[~to_add['Auid'].isna()]
```

```python
new_site_name = 'WCHRI'
all_sites = ['ACHRI', 'BCCHR', 'CHRIM', 'CHUSJ', 'MCMASTER', 'OBRIEN', 'Sickkids', 'WCHRI']
all_sites.remove(new_site_name)

# create a df of just Auids and WCHRI indicator
to_add[new_site_name] = 1
for site in all_sites:
    to_add[site] = 0
to_add.drop('Name', axis=1, inplace=True)

# remove WCHRI from df_chw
df_chw = df_chw[df_chw['WCHRI'] == 0]

# replace the WCHRI Auids with the updated ones
df_chw = pd.concat([to_add, df_chw])
df_chw
```

```python
df_chw = df_chw.reset_index(drop=True)
df_chw
```

```python
df_chw.info()
```

```python
df_chw['Auid'] = df_chw['Auid'].apply(int)
```

```python
df_chw.to_excel('../../Datasets/chw_ids.xlsx')
```
